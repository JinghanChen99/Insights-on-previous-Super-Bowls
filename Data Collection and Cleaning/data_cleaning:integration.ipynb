```python
import pandas as pd
import numpy as np
```

```python
needed_col=['Date','SB','Attendance','QB  Winner','Coach Winner','Winner','Winning Pts','QB Loser','Coach Loser',\
            'Loser','Losing Pts','MVP','Stadium','City','State','Point Difference']
base_df = pd.read_csv('Data Collection/SuperBowl_base_update.csv', index_col='SB', usecols=needed_col)
base_df
```

```python
base_df['Coach Winner'] = base_df['Coach Winner'].str.replace('_','')
base_df['Coach Loser'] = base_df['Coach Loser'].str.replace('_','')
base_df.index
```

```python
base_df.dtypes
```

```python
base_df.loc['LIII', 'Attendance'] = base_df.loc['LIII', 'Attendance'].replace(',','')
base_df.loc['LIV', 'Attendance'] = base_df.loc['LIV', 'Attendance'].replace(',','')
base_df.Attendance.astype('int64')
```

```python
envir_df = pd.read_csv('Data Collection/envir.csv')
envir_df.head()
```

```python
viewer_df = pd.read_csv('Data Collection/viewers_Ads_data.csv', usecols=['Average Viewers','Total Viewers','TV_rating','TV_share','30-sec Ads cost'])
viewer_df.head()
```

```python
competitiveness_df = pd.read_csv('Data Collection/competitiveness.csv', usecols=['competitive_score'])
competitiveness_df.head()
```

```python
winner_df = pd.read_csv('Data Collection/Winner_season_stats_complete.csv', usecols=['winnor_season_winning_pct','winnor_season_point_score',\
                                                                     'winnor_season_point_allow','winnor_end_streak'])
winner_df.head()
```

```python
winner_df.dtypes
```

```python
winner_df['Winning_streak'] = np.where(winner_df['winnor_end_streak'].str.get(0) == "W", winner_df['winnor_end_streak'].str[1:],0)
winner_df['Winning_streak'] = winner_df['Winning_streak'].astype('int64')
winner_df['Losing_streak'] = np.where(winner_df['winnor_end_streak'].str.get(0) == "L", winner_df['winnor_end_streak'].str[1:],0)
winner_df['Losing_streak'] = winner_df['Losing_streak'].astype('int64')
winner_df.head()
# winner_df.dtypes
```

```python
loser_df  = pd.read_csv('Data Collection/loser_season_stats _complete.csv',usecols=['loser_season_winning_pct','loser_season_point_score',\
                                                                     'loser_season_point_allow','loser_end_streak'])
loser_df.head()
```

```python
loser_df.dtypes
```

```python
loser_df['Winning_streak'] = np.where(loser_df['loser_end_streak'].str.get(0) == "W", loser_df['loser_end_streak'].str[1:],0)
loser_df['Winning_streak'] = loser_df['Winning_streak'].astype('int64')
loser_df['Losing_streak'] = np.where(loser_df['loser_end_streak'].str.get(0) == "L", loser_df['loser_end_streak'].str[1:],0)
loser_df['Losing_streak'] = loser_df['Losing_streak'].astype('int64')
loser_df.head()
# winner_df.dtypes
```

```python
new_info = pd.concat([envir_df, viewer_df, competitiveness_df, winner_df, loser_df], axis=1)
new_info.head()
```

```python
base_df = base_df.reset_index()
```

```python
final_dataset = pd.concat([base_df, new_info], axis=1).set_index('SB')
final_dataset.columns
```
